```python
data.replace(('M','B'),(1,0), inplace=True)
new_data = data.drop(['ID'],axis=1)
new_data.head()

labels= data.iloc[:,1]
labels
```

# Exercise 3

```python
# you may need to run  pip install sklearn in Ubuntu  (may need pip3)
from sklearn.decomposition import PCA
import matplotlib
from matplotlib import pyplot as plt


x = np.array(new_data)
print(np.shape(x))

y = np.array(labels)
print(np.shape(y))
```

```python

plt.figure()
plt.plot(x[y==1,2],x[y==1,3],'b.')
plt.plot(x[y==0,2],x[y==0,3],'g.')
plt.show()
```

```python
plt.figure()
plt.plot(x[y==1,8],x[y==1,1],'b.')
plt.plot(x[y==0,8],x[y==0,1],'g.')
plt.show()
```

```python
plt.figure()
plt.plot(x[y==1,12],x[y==1,3],'b.')
plt.plot(x[y==0,12],x[y==0,3],'g.')
plt.show()
```

```python
plt.figure()
plt.plot(x[y==0,3],x[y==0,12],'b.')
plt.plot(x[y==1,3],x[y==1,12],'g.')
plt.show()
```

```python
plt.figure()
plt.plot(x[y==1,30],x[y==1,3],'b.')
plt.plot(x[y==0,30],x[y==0,3],'g.')
plt.show()
```

```python
plt.figure()
PCA3=PCA(n_components=2)
XPCA = PCA3.fit_transform(x)
plt.plot(XPCA[y==0,0],XPCA[y==0,1],'g.')
plt.plot(XPCA[y==1,0],XPCA[y==1,1],'r.')
plt.show()
```


the benign (green) dots are condensing in a particular range compared to the scatter spread that is common in previous graphs.
