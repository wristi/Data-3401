## Problem 2
1. For a given variable num_trials, generate a 100 by num_trials ndarray with entries from the standard normal distribution (look up the syntax for numpy.random.randn).

1. For each of the following values of num_trials 

1. {10,100,1000,10000,1000000}, generate the ndarray above.

1. For each value of num_trials, sum the columns of the ndarray (you should get a new ndarray of length num_trials).

1. Use pyplot to plot a histogram of the summed array for each value of num_trials and various numbers of bins.

1. Discuss any trends you see.

```python
import matplotlib.pyplot as plt
def bellcurve(num_trials):
    A = np.random.randn(100,num_trials)
    SUM = np.sum(A,axis=0)
    #print(SUM)
    plt.hist(SUM,bins = 10, range = (-3,3))
    plt.show()
```

```python
bellcurve(100)
#Anytime there is a drastic high point,there are holes in the graph
#Just enough data to give a traditional bell curve
```

```python
bellcurve(1000)
#Just enough Data to shape out the graph, with some c
```

```python
bellcurve(10000)
```

```python
bellcurve(100000)
```

```python
a = np.random.randint(low = 1, high = 11, size=(15))
print(type(a))
print(a)

A = np.random.choice(a, size=None, replace=True, p=None)
print(A)
#When replace is on, elements of the array can be selected many times.
```

```python
a = np.random.randint(low = 1, high = 11, size=(15))
print(a)
A = np.random.choice(a, size=None, replace=False, p=None)
print(A)
#When replace is off, array elements can only be selected once.
```

```python
b = np.random.randint(low = 1, high = 11, size=(4))
print(b)
probs=[.8, .1, .1, 0]  
B = np.random.choice(b, size=None, replace=True, p=probs)
print(B)
#since p[0] has the highest probabiltiy, it is almost always picked 
```

```python
c = np.random.randint(low = 1, high = 21, size=(4))
print(c)

C = np.random.choice(c, size=None, replace=True, p=probs)

print(C)
EMean = np.sum(c, dtype=np.float64)

print(c/EMean * 100)
```

```python
d = np.random.randint(low = 1, high = 101, size=(4))
print(d)

D = np.random.choice(d, size=None, replace=True, p=probs)

print(D)
DMean = np.sum(d, dtype=np.float64)
print(d/DMean * 100)
totalthis = d/np.sum(d)
np.sum(totalthis)
```
